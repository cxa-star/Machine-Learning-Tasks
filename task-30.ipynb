```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Walmart_Sales.csv")
```

```python
plt.figure(figsize=(15,10))
```

```python
plt.subplot(2,2,1)
plt.plot(df["Weekly_Sales"])
plt.title("Weekly Sales Trend")
```

```python
plt.subplot(2,2,2)
df.groupby("Store")["Weekly_Sales"].sum().plot(kind="bar")
plt.title("Store-wise Sales")
```

```python
plt.subplot(2,2,3)
plt.scatter(df["CPI"], df["Weekly_Sales"])
plt.title("CPI vs Sales")
```

```python
plt.subplot(2,2,4)
plt.scatter(df["Unemployment"], df["Weekly_Sales"])
plt.title("Unemployment vs Sales")
```

```python
plt.tight_layout()
plt.show()
```
